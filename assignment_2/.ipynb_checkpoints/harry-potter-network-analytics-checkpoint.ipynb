```python
import matplotlib.pyplot as plt
import sys
import pandas as pd
import networkx as nx
from operator import itemgetter
from itertools import combinations
import re
```

```python
file = open('harry_potter.txt','rb')
```

```python
text = file.read().decode(encoding='utf-8',errors='replace')

text = text.replace('\n',' ')
text = text.replace('\'','')
text = text.replace('-',' ')
text = text.lower()

def normalcharacters(x):
    for i in x:
        if(i<'a' or x>'z'):
            return False
    return True

text = ' '.join([x for x in text.split() if(normalcharacters(x))])
print(text)
```

```python
words_list = text.split()
```

```python
len(words_list)
```

```python
characters = ["Harry", "Hermione", "Ron", "Dursleys", "Dumbledore", "Voldemort", "Hagrid", "Draco", "McGonagall", "Snape", "Quirrell", "Neville", "Vernon", "Petunia", "Dudley", "Hedwig", "Ollivander", "Fred","George", "James","Lily","Molly","Percy","Filch","Nicolas","Oliver","Lee","Seamus","Charlie","Vincent","Gregory","Hooch","Fang","Fluffy","Griphook","Malkin","Cornelius","Lavender","Pomfrey","Katie","Angelina","Marcus","Irma","Alicia","Flitwick","Sprout","Binns","Piers","Augusta","Trevor","Ginny","Gellert","Tom","Sirius"]
```

```python
characters = [x.lower() for x in characters]
```

```python
print(characters)
len(characters)
```

```python
links = []
window = 10
window_parts = []
for k in range(0,len(words_list),window):
    text_current = words_list[k:k+window]
    window_parts.append(' '.join(text_current))

print(window_parts)
for i in range(0,len(characters)):
    for j in range(i+1,len(characters)):
        count = 0
        for part in window_parts:
            if (len(re.findall(characters[i],part)) > 0) and (len(re.findall(characters[j],part)) > 0):
                count += 1
        edge = []
        edge.append(characters[i])
        edge.append(characters[j])
        edge.append(count)
        links.append(edge)        
```

```python
mygraph = nx.Graph()
mygraph.add_nodes_from(characters)
print(links)
for i in links:
    if(i[2]!=0):
        mygraph.add_edge(i[0],i[1],weight=int(i[2]))
print(nx.info(mygraph))
```

```python
print('Question A')
plt.figure(3,figsize=(16,16)) 
nx.draw_networkx(mygraph,node_color='yellow',with_labels=True)
plt.show()
```

```python
print('Question D')
print('Pearson Correlation coefficient')
print(nx.degree_pearson_correlation_coefficient(mygraph))

cc=nx.closeness_centrality(mygraph)
dc=nx.degree_centrality(mygraph)
bc = nx.betweenness_centrality(mygraph)
ec = nx.eigenvector_centrality(mygraph,max_iter=1000)
sorted_dc=sorted(dc.items(),reverse=True, key = lambda x: x[1])
sorted_bc=sorted(bc.items(),reverse=True, key = lambda x: x[1])
sorted_cc=sorted(cc.items(),reverse=True, key = lambda x: x[1])
sorted_ec=sorted(ec.items(),reverse=True, key = lambda x: x[1])
print(sorted_dc)
print(sorted_bc)
print(sorted_cc)
print(sorted_ec)

deg_data=[y for x,y in sorted_dc]
close_data=[y for x,y in sorted_cc]
betw_data=[y for x,y in sorted_bc]
eig_data=[y for x,y in sorted_ec]

from scipy.stats.stats import pearsonr
r1 = pearsonr(deg_data,betw_data)
s = "\n Pearson Correlation Coeff deg and betweenness: " + repr(r1)
print(s)
r2 = pearsonr(deg_data,close_data)
s = "\n Pearson Correlation Coeff deg and closeness: " + repr(r2)
print(s)
r3 = pearsonr(betw_data,close_data)
s = "\n Pearson Correlation Coeff betweennness and closeness: " + repr(r3)
print(s)
r4 = pearsonr(betw_data,eig_data)
s = "\n Pearson Correlation Coeff betweennness and eigenvector : " + repr(r4)
print(s)
```

```python
print('Question E')
import collections

def degree_histogram(G):
    degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
    # print "Degree sequence", degree_sequence
    degreeCount = collections.Counter(degree_sequence)
    deg, cnt = zip(*degreeCount.items())
    fig, ax = plt.subplots()
    plt.bar(deg, cnt, width=0.80, color='b')

    plt.title("Degree Histogram")
    plt.ylabel("Count")
    plt.xlabel("Degree")
    ax.set_xticks([d + 0.4 for d in deg])
    ax.set_xticklabels(deg)

    # draw graph in inset
    plt.axes([0.4, 0.4, 0.5, 0.5])
    Gcc = sorted(nx.connected_component_subgraphs(G), key=len, reverse=True)[0]
    pos = nx.spring_layout(G)
    plt.axis('off')
    #nx.draw_networkx_nodes(G, pos, node_size=20)
    #nx.draw_networkx_edges(G, pos, alpha=0.4)

    plt.show()

    
print('Degree Distribution Histogram')
degree_histogram(mygraph)

degree_sequence = sorted([d for n, d in mygraph.degree()], reverse=True)  # degree sequence
print(degree_sequence)
print('Log Log Plot')
plt.loglog(degree_sequence,'b-',marker='o')
plt.title("Degree rank plot")
plt.ylabel("degree")
plt.xlabel("rank")
```

```python
print('Question O')
print('Detect community using Louvain and Clique percolation methods\n')
from networkx.algorithms.community import k_clique_communities
import community

print('Louvain Method')
partition = community.best_partition(mygraph)
print(partition)
#drawing
size = float(len(set(partition.values())))
pos = nx.spring_layout(mygraph)
count = 0.
for com in set(partition.values()) :
    count = count + 1.
    list_nodes = [nodes for nodes in partition.keys()
                                if partition[nodes] == com]
    nx.draw_networkx_nodes(mygraph, pos, list_nodes, node_size = 20,
                                node_color = str(count / size))


nx.draw_networkx_edges(mygraph, pos, alpha=0.5)
plt.show()


print('Clique Percolation Method')
print(list(k_clique_communities(mygraph,3)))
```

<h2> Reason for disconnected graph </h2>

In Harry Potter novels, there are a lot of characters who are only mentioned in the third person as a charcter of background interest and as a desciption of some place of importance. These characters have no connection with any other character in the book. Examples of such characters include marcus, katie, irma. Also we have used a window size of 10. If we had used a higher window size, we could have forced some connections involving these nodes. 

<h2> a) Centrality Analysis </h2>

<ul>
<li>Harry: The main protagonist of the story "Harry Potter and the Sorceror's Stone". This explains why he has the highest degree centrality (connected to most of the characters), betweenness centrality (acting as a bridging node between different communities), closeness centrality (being at the centre of the whole network) and eigenvector centrality (connecting extremely influential characters in the story). The story revolves around him, as is evident from the title. He is the centre of the network, and directly interacts with most of the other characters.</li>
<li>Ron: One of the two companions of Harry participating heavily in the storyline. This explains the high centrality scores for this character. Being born in a wizarding family and his whole family being a part of the storyline, Ron does have more connections than other characters.</li>
<li>Hermione: One of the two companions of Harry participating heavily in the storyline. This explains the high centrality scores for this character. Hermione was born in a non-wizarding family and this explains the lesser centrality scores when compared to Ron.</li>
<li>Voldemort: The evil character appearing in the story. The scores do not actually show the importance of this character because this character was referenced using many aliases like "The Dark Lord", "He who must not be named", etc.</li>
</ul>

<p>Here, we see that the graph corresponding to Harry Potter and the Sorceror's Stone has an abnormally large number of 0-degree nodes when compared with the small_world graph. Also, the degrees are mostly concentrated among the top 4 characters in the graph corresponding to the novel whereas in the small world graph, the degree distribution roughly follows a bell curve. The graph corresponding to the novel also has many disconnected components while the small world graph has a single component with a diameter of 4. Thus, we can safely assume that the network as described in the novel "Harry Potter and the Sorceror's Stone" is far from a real network.</p>

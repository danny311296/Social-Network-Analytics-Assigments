```python
%matplotlib inline
import matplotlib.pyplot as plt
import networkx as nx
import operator
mygraph = nx.read_gml("lesmis/lesmis.gml")
```

```python
print(nx.info(mygraph))
```

```python
print('Question A')
nx.draw_networkx(mygraph)
```

```python
print('Question B')
print('\nNodes')
print(mygraph.nodes)
print('\nEdges')
print(mygraph.edges)
print('\nEdgesWithAttributes')
print(nx.get_edge_attributes(mygraph,'value'))
```

```python
print('Question C')
print('\nDegreeCentrality')
dc=nx.degree_centrality(mygraph)
print(sorted(list(dc.items()),key=operator.itemgetter(1),reverse=True))
print('\nClosenessCentrality')
cc=nx.closeness_centrality(mygraph)
print(sorted(list(cc.items()),key=operator.itemgetter(1),reverse=True))
print('\nBetweennessCentrality')
bc=nx.betweenness_centrality(mygraph)
print(sorted(list(bc.items()),key=operator.itemgetter(1),reverse=True))
print('\nEigenvectorCentrality')
ec=nx.eigenvector_centrality(mygraph)
print(sorted(list(ec.items()),key=operator.itemgetter(1),reverse=True))
```

```python
print('Question D')
print(nx.degree_pearson_correlation_coefficient(mygraph))

cc=nx.closeness_centrality(mygraph)
dc=nx.degree_centrality(mygraph)
bc = nx.betweenness_centrality(mygraph)
ec = nx.eigenvector_centrality(mygraph,max_iter=1000)
sorted_dc=sorted(dc.items())
sorted_bc=sorted(bc.items())
sorted_cc=sorted(cc.items())
sorted_ec=sorted(ec.items())

deg_data=[y for x,y in sorted_dc]
close_data=[y for x,y in sorted_cc]
betw_data=[y for x,y in sorted_bc]
eig_data=[y for x,y in sorted_ec]

from scipy.stats.stats import pearsonr
r1 = pearsonr(deg_data,betw_data)
s = "\n Pearson Correlation Coeff deg and betweenness: " + repr(r1)
print(s)
r2 = pearsonr(deg_data,close_data)
s = "\n Pearson Correlation Coeff deg and closeness: " + repr(r2)
print(s)
r3 = pearsonr(betw_data,close_data)
s = "\n Pearson Correlation Coeff betweennness and closeness: " + repr(r3)
print(s)
r4 = pearsonr(betw_data,eig_data)
s = "\n Pearson Correlation Coeff betweennness and eigenvector : " + repr(r4)
print(s)
```

```python
print('Question E')
import collections

def degree_histogram(G):
    degree_sequence = sorted([d for n, d in G.degree()], reverse=True)  # degree sequence
    # print "Degree sequence", degree_sequence
    degreeCount = collections.Counter(degree_sequence)
    deg, cnt = zip(*degreeCount.items())
    fig, ax = plt.subplots()
    plt.bar(deg, cnt, width=0.80, color='b')

    plt.title("Degree Histogram")
    plt.ylabel("Count")
    plt.xlabel("Degree")
    ax.set_xticks([d + 0.4 for d in deg])
    ax.set_xticklabels(deg)

    # draw graph in inset
    plt.axes([0.4, 0.4, 0.5, 0.5])
    Gcc = sorted(nx.connected_component_subgraphs(G), key=len, reverse=True)[0]
    pos = nx.spring_layout(G)
    plt.axis('off')
    #nx.draw_networkx_nodes(G, pos, node_size=20)
    #nx.draw_networkx_edges(G, pos, alpha=0.4)

    plt.show()

degree_histogram(mygraph)

degree_sequence = sorted([d for n, d in mygraph.degree()], reverse=True)  # degree sequence
print(degree_sequence)
plt.loglog(degree_sequence,'b-',marker='o')
plt.title("Degree rank plot")
plt.ylabel("degree")
plt.xlabel("rank")
```

```python
print('Question K')
a = nx.degree_centrality(mygraph)
#print(a)
sorted_by_value = sorted(a.items(), key=lambda kv: kv[1],reverse=True)
print(sorted_by_value[:3])
for i,j in sorted_by_value[:3]:
    ego = nx.ego_graph(mygraph,i)
    pos = nx.spring_layout(ego)
    nx.draw(ego,pos,node_color='b',node_size=50,with_labels=True)
    nx.draw_networkx_nodes(ego,pos,nodelist=[i],node_size=300,node_color='r')
    plt.show()
    print('Alters: ')
    for k in mygraph.neighbors(i):
        print(k + ' ' , end='')
    
```

```python
print('Question L')
a = nx.jaccard_coefficient(mygraph, ebunch=None)
lst=[]
for i,j,k in a:
    lst.append([i,j,k])
lst.sort(key=operator.itemgetter(2),reverse=True)
print(lst[:5])
print('\nAll:\n',lst)
```

```python
print('Question M')
print(nx.clustering(mygraph))
print('\nGlobalClusteringCoeff.')
print(nx.average_clustering(mygraph))
```

```python
print('Question N')  
print('Detect cliques and Cores')

print('\nMaximal Cliques')
a = list(nx.find_cliques(mygraph))
print(a)

print('\nCores')
for i in range(2,10):
    print(str(i) + ' Core')
    cores = nx.k_core(mygraph,i)
    nx.draw_networkx(cores,node_color='cyan')
    plt.show()
```

```python
print('Question O')
import community
from networkx.algorithms.community.centrality import girvan_newman
communities_generator = girvan_newman(mygraph)
top_level_communities = next(communities_generator)
next_level_communities = next(communities_generator)
sorted(map(sorted, next_level_communities))
part = community.best_partition(mygraph)
print(community.modularity(part, mygraph))
induced = community.induced_graph(part, mygraph)
print(induced.nodes())
print(induced.edges())
nx.draw_networkx(induced)
plt.show()
from networkx.algorithms.community import k_clique_communities
print(list(k_clique_communities(mygraph,4))) # Try different clique sizes also

print('Louvain Method')
partition = community.best_partition(mygraph)
print(partition)
#drawing
size = float(len(set(partition.values())))
pos = nx.spring_layout(mygraph)
count = 0.
for com in set(partition.values()) :
    count = count + 1.
    list_nodes = [nodes for nodes in partition.keys()
                                if partition[nodes] == com]
    nx.draw_networkx_nodes(mygraph, pos, list_nodes, node_size = 20,
                                node_color = str(count / size))


nx.draw_networkx_edges(mygraph, pos, alpha=0.5)
plt.show()


print('Clique Percolation Method')
print(list(k_clique_communities(mygraph,3)))
```

```python
import seaborn as sns
import pandas as pd
```

```python
centrality_measures = {
'degree': dc,
'betweenness': bc,
'closeness': cc,
'eigenvector': ec,
}
centrality = pd.DataFrame(centrality_measures)
sns.pairplot(centrality)
```

<h4>Analysis(a)</h4>

<ul>
<li>Jean Valjean: The main protagonist of the story "Les Miserables". This explains why he has the highest degree centrality (connected to most of the characters), betweenness centrality (acting as a bridging node between different communities) and closeness centrality (being at the centre of the whole network).</li>
<li>Gavroche: A member of the "Friends of the ABC" who had to deliver the letter from Marius to Cosette, Valjean's adopted daughter. This character has the highest eigenvector centrality as it is connecting two extremely influential people in the story.</li>
<li>Marius: A young character who falls in love with Cosette, Valjean's adopted daughter. Scores high centrality values as most of the later part of the story mainly revolves aroung the affair between him and Cosette.</li>
<li>Javert: The police chief of Montreul, who then appears in Paris too. This character was primarily connected only to Valjean, thus explaining the comparatively lower eigenvector and betweenness centralities.</li>
</ul>

Some of the communities detected were:
<ul>

<li>['Anzelma', 'MmeThenardier', 'Thenardier', 'Eponine']: The Thenardier family, including the two daughters Eponine and Azelma.</li>
<li>['Valjean', 'Javert', 'Thenardier', 'Cosette', 'MmeThenardier']: The protagonist Valjean rescues Cosette by buying her from the Thenardiers and then moving to Paris, but Javert discovers their hideout. These characters interacte repeatedly.</li>
</ul>
